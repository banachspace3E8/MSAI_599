```python
#pip install kagglehub
#One time run
import kagglehub

# Download latest version (code obtained from: https://www.kaggle.com/datasets/rabieelkharoua/predict-restaurant-customer-satisfaction-dataset?resource=download)
path = kagglehub.dataset_download("rabieelkharoua/predict-restaurant-customer-satisfaction-dataset")
print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

full_path = os.path.join(path, "restaurant_customer_satisfaction.csv")
df = pd.read_csv(full_path)
print(df.head())
```

```python
#(#_of_rows, #_of_cols)
df.shape
```

```python
df.info()
```

```python
#Basic statistical properties of numerical dataset
df.describe()
```

```python
# Some of these analysis are not relevant for the categorical dataset
#'unique' feature shows how many subcategories do each categorical ddataset has
df.describe(include='all')
```

```python
#Identify if there is any missing entries
df.isnull().sum()
```

```python
print(df['Gender'].unique())
print(df['VisitFrequency'].unique())
print(df['PreferredCuisine'].unique())
print(df['TimeOfVisit'].unique())
print(df['DiningOccasion'].unique())
print(df['MealType'].unique())
```

```python
#Histogram of counts of Categorical Dataset
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 4, figsize=(9, 9), constrained_layout=True)

df['Gender'].value_counts().plot(kind='bar', title=f'Counts of Gender', ax=axes[0,0]);
df['VisitFrequency'].value_counts().plot(kind='bar', title=f'Counts of Visit Frequency', ax=axes[0,1]);
df['PreferredCuisine'].value_counts().plot(kind='bar', title=f'Counts of Preferred Cuisine', ax=axes[0,2]);
df['ServiceRating'].value_counts().plot(kind='bar', title=f'Counts of Service Rating', ax=axes[0,3]);
df['TimeOfVisit'].value_counts().plot(kind='bar', title=f'Counts of Time Of Visit', ax=axes[1,0]);
df['DiningOccasion'].value_counts().plot(kind='bar', title=f'Counts of Dining Occasion', ax=axes[1,1]);
df['MealType'].value_counts().plot(kind='bar', title=f'Counts of Meal Type', ax=axes[1,2]);
df['GroupSize'].value_counts().plot(kind='bar', title=f'Counts of Group Size', ax=axes[1,3]);
```

```python
#Histogram of counts of Numerical Dataset
#10 bins
counts, bin_edges, patches = plt.hist(df['Age'], bins=10, color='green', edgecolor='black', alpha=0.9)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Age Distribution Histogram in 10 equal bins', fontsize=14)
print(bin_edges)
```
